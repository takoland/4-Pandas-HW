```python
import pandas as pd
```

```python
file_1 = "PyCitySchools/raw_data/students_complete.csv"
students_df = pd.read_csv (file_1)
students_df.head()
```

```python
file_2 = "PyCitySchools/raw_data/schools_complete.csv"
schools = pd.read_csv (file_2)
schools.head()
```

```python
schools_df = schools.rename(columns = {"name": "school"})
schools_df.head()
```

```python
merged_pd = pd.merge (students_df, schools_df, on="school",how="inner")
merged_pd.head()
```

```python
merged_pd ["type"].value_counts()
#school_grouped = reorder_df.groupby(["school"])
#school_grouped.mean()
```

```python
#district summary
#total school
total_district = merged_pd ["school"].value_counts().count()
#total students 
total_students = merged_pd ["Student ID"].count()
#total budget 
district_budget = merged_pd [["budget","school"]]
school_budget = district_budget.groupby(["school"])
#grouped_usa = usa_ufo.groupby(['state'])
total_school_budget = school_budget.mean()
total_budget = total_school_budget.sum()

#average math score
avg_math = merged_pd ["math_score"].mean()
#average reading score
avg_read = merged_pd ["reading_score"].mean()
#%passing math
#Only_Billys_and_peter = df.loc [ (df [ "first_name" ] == "Billy" ) | (df [ "first_name"] == "Peter") , : ] 
#Only_billys = df.loc [ df [ " first_name" ] == "Biily" , : ] 
name_pass_math = merged_pd .loc[merged_pd ["math_score"] > 70 ,:]
number_pass_math = name_pass_math["Student ID"].count()
percent_pass_math = (number_pass_math / total_students)*100
# % passing reading
name_pass_read = merged_pd .loc[merged_pd ["reading_score"] > 70 ,:]
number_pass_read = name_pass_read["Student ID"].count()
percent_pass_read = (number_pass_read / total_students)*100
# % overall passing rate 
overall_percent =((percent_pass_math + percent_pass_read ) / 200)*100
```

```python
district_summary = pd.DataFrame ( {"Total Schools":[total_district],
                                    "Total Students":[total_students],
                                      "Total Budget":[int(total_budget)],
                                      "Average Math Score":[avg_math],
                                      "Average Reading Score":[avg_read],
                                      "% Passing Math":[percent_pass_math],
                                      "% Passing Reading":[percent_pass_read],
                                      "% Overall Passing Rate":[overall_percent]})

district_summary["Total Students"]= district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("{:,}".format)
district_summary["% Overall Passing Rate"] = district_summary["% Overall Passing Rate"].map("{:.2f}".format)
district_summary["% Passing Reading"] = district_summary["% Passing Reading"].map("{:.2f}".format)
district_summary["% Passing Math"] = district_summary["% Passing Math"].map("{:.2f}".format)
district_summary["Average Math Score"] = district_summary["Average Math Score"].map("{:.2f}".format)
district_summary["Average Reading Score"] = district_summary["Average Reading Score"].map("{:.2f}".format)
district_summary
```
